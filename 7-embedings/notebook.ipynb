```python
from gensim.models import KeyedVectors
from pprint import pprint
from sklearn.manifold import TSNE
import numpy as np
import matplotlib.pyplot as plt
model = KeyedVectors.load_word2vec_format("./skipgram/skip_gram_v100m8.w2v.txt", binary=False)
print("Loaded data")
```

```python

def append_noun(word):
    return word + "::noun"

lemmas = [
    "ne#Trybunał_konstytucyjny",
    "kodeks_cywilny",
    "kpk",
    "sąd_rejonowy",
    "szkoda",
    "wypadek",
    "kolizja",
    "szkoda_majątkowa",
    "nieszczęście",
    "rozwód",
    "sąd_wysoki",
]

for lemma in lemmas:
    try:
        most_similar = model.most_similar(positive=append_noun(lemma))
        pprint(lemma)
        pprint(most_similar)
    except Exception as e:
        print(e)

high_judge_phrase = ["sąd::noun", "wysoki::adj"]
most_similar = model.most_similar(positive=high_judge_phrase)
pprint(high_judge_phrase)
pprint(most_similar)
```

```python

equations =[
    (["sąd wysoki", "konstytucja"],["kpc"]),
    (["pasażer", "kobieta"], ["mężczyzna"]),
    (["samochód", "rzeka"], ["droga"])
]
for equation in equations:
    positives = [append_noun(w) for w in equation[0]]
    negatives = [append_noun(w) for w in equation[1]]

    try:
        print(equation)
        most_similar = model.most_similar(positive=positives, negative=negatives)
        pprint(most_similar)
    except Exception as e:
        print(e)
        
high_judge_phrase = ["sąd::noun", "wysoki::adj"]
most_similar = model.most_similar(positive=high_judge_phrase, negative=["kpc::noun"])
pprint(high_judge_phrase)
pprint(most_similar)

    
```

```python

words = [
    "szkoda",
    "strata",
    "uszczerbek",
    "szkoda_majątkowa",
    # "krzywda", - not existing - only Surnames
    #"uszczerbek na zdrowie", - only uszczerbek exists
    "niesprawiedliwość",
    "nieszczęście",
]
vectors =[]
for word in words:
    phrase = append_noun(word)
    result = model.get_vector(phrase)
    vectors.append(result)
vectors = np.array(vectors)
```

```python
from random import sample

vocab=list(model.vocab)
random_data = sample(vocab, 1000)
random_vectors = np.array([model.get_vector(v) for v in random_data])
```

```python
X = np.concatenate((random_vectors, vectors))
```

```python
result = TSNE().fit_transform(X)
```

```python
print(result)
plt.figure(figsize=(10, 10))

plt.scatter(result[:, 0], result[:, 1], c=np.array(["blue" if x < 1000 else "red" for x in range(len(result))]))
plt.show()
```
